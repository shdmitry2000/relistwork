load file from excell

```python
import pandas as pd

df = pd.read_excel (r'Cin-Leads - Podio export - July-16.xlsx') #for an earlier version of Excel, you may need to use the file extension of 'xls'
print (df)
```

get only new leads

```python
newleadsDF = df.loc[df['Status'] == 'New']

newinprogressleadsDF=newleadsDF.loc[newleadsDF['Lead Phase'] == 'In Progress']
print(newinprogressleadsDF)
```

drop unused colums 

what are columns names

```python
print(len(newinprogressleadsDF.columns))
print(newinprogressleadsDF.columns)
```

drop

```python
newcleandf= newinprogressleadsDF.drop(newinprogressleadsDF.columns[15:69], axis=1)
print(newcleandf.columns)

print(newcleandf.head)
```

```python
print(len(newcleandf.columns))
```

show duplicate rows

```python
# Select all duplicate rows based on one column
duplicateRowsDF = newcleandf[newcleandf.duplicated(['Lead Reference'])]
print("we found",duplicateRowsDF.size , "rows")
duplicateRowsDF
```

we drop all duplicates 
leave last one

```python
# Select all duplicate rows based on one column keep only last
woduplicatesRowsDF = newcleandf.drop_duplicates(['Lead Reference'], keep='last')
print("we found",woduplicatesRowsDF.size , "rows")
woduplicatesRowsDF
```
